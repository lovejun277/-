```python
import pandas as pd

# 读取Excel文件并将所有表格存储在一个字典中
excel_file = pd.ExcelFile(r'D:\新建文件夹\2020.1-2022.12工资(1).xlsx')
dfs = {}

for sheet_name in excel_file.sheet_names:
    df = excel_file.parse(sheet_name)
    # 添加一列以记录sheet_name
    df['Sheet Name'] = sheet_name
    dfs[sheet_name] = df

# 使用 pd.concat() 合并所有表格
merged_df = pd.concat(dfs.values(), ignore_index=True)

# 输出合并后的DataFrame
print(merged_df)

```
