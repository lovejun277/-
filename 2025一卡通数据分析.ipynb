```python
import pandas as pd

# 假设df是你的DataFrame
# 如果还未读取数据，可以使用: df = pd.read_excel('your_file.xlsx')

# 按补贴类型分组并汇总
summary = df.groupby('补贴类型').agg({
    '补贴数量': 'sum',
    '应发金额': 'sum',
    '实发金额': 'sum'
}).reset_index()

# 显示汇总结果
print(summary)

# 可选：保存汇总结果到Excel
summary.to_excel('补贴类型汇总.xlsx', index=False)
```

```python
#df3 = pd.merge(df1, df2, on='A', how='left')
#除了 left 之外，pandas.merge() 函数的 how 参数还可以取以下值：

#'inner': 内连接，仅返回两个表中都存在的记录。
#'right': 右连接，返回右表中的所有记录，以及左表中连接键匹配的记录。
#'outer': 外连接，返回两个表中的所有记录，以及连接键匹配的记录。
df3 = pd.merge(df,df2, left_on='身份证号', right_on='证件号码', how='inner', indicator=True)
show(df3)
```
